```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("tmdb_5000_credits.csv")
```

```python
credits = pd.read_csv("tmdb_5000_movies.csv")
```

```python
movies.head()
```

```python
credits.head()
```

```python
 movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
movies.shape
```

```python
#colums we are going to keep
# genres , id , keywords, title , overview , casts, crew,
```

```python
movies = movies[['genres','keywords','title','overview','cast','crew','id']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
def converter(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
import ast
```

```python
movies['genres'] = movies['genres'].apply(converter)
```

```python
movies.head()['genres']
```

```python
movies['keywords'] = movies['keywords'].apply(converter)
```

```python
#we only want the names of three leadiing actors 
def converter3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3 :
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
        
```

```python
movies['cast'] = movies['cast'].apply(converter3)
```

```python
movies.head()
```

```python
 movies['crew'][0]
```

```python
def job(obj):
    L  = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            L.append(i["name"])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(job)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply( lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply( lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply( lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply( lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tag']= movies['overview']+movies['keywords']+movies['genres']+movies['cast']
```

```python
new_df = movies[['id','title','tag']]
```

```python
new_df.head()
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:" ".join(x))
new_df['tag'] = new_df['tag'].apply(lambda x:x.lower().strip())
```

```python
new_df.head()
```

```python
#text vectorizationn :
#coverting text into vectors now we will recommend the closest to the respective vector , here we will use a tecniqus "bag of words"
# we will concat all the tags and we will count their frequency and extractig the one which have the maximun frequency
# removing stop words and using scikit learn to remove thoe
```

```python
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
#converting the matrix in numpy arrays
```

```python
vectors[0]
```

```python
cv.get_feature_names()
#getting the words according to their frequency
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join   
```

```python
new_df['tag'] = new_df['tag'].apply(stem)
```

```python
#to measure te distance we will not use eccledian distance we will use the angle btween those two vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
 new_df[new_df['title'] == 'Avatar']
```

```python
#*****recommend function******
def recommend(movie):
    movies_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movies_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key= lambda x : x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
    return
```

```python
recommend("Batman")
```
